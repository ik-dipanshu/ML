```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.DataFrame({'exp':[1,2,3,4,5],'salary':[50,20,60,40,70]})
plt.scatter(df['exp'],df['salary'])
plt.show()
x=df['exp']
y=df['salary']
```

```python
model=LinearRegression()
model.fit(x_train_poly,y_train)
```

```python
import pandas as pd
import numpy as np

np.random.seed(42)

experience = 5*np.random.rand(200,1)

salary =0.6*(experience)**2+0.8*experience+3+np.random.randn(200,1)

df=pd.DataFrame(np.hstack((experience,salary)),columns=['Experience','Salary'])
```

```python
x=df['Experience']
y=df['Salary']
df
```

```python
plt.scatter(x,y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train.shape #160
x_test.shape #40
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=2,include_bias=True)
x_train_poly=poly.fit_transform(x_train.values.reshape(-1,1))
x_test_poly=poly.fit_transform(x_test.values.reshape(-1,1))
x_train_poly
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train_poly,y_train)
predict=model.predict(x_test_poly)
```

```python
plt.scatter(x_train,y_train,color='red',label='training')
plt.scatter(x_test,y_test,color='blue',label='testing')

# Sort x_test and corresponding predict values to draw a smooth line
sorted_indices = x_test.argsort()
sorted_x_test = x_test.iloc[sorted_indices]
sorted_predict = predict[sorted_indices]

plt.plot(sorted_x_test,sorted_predict,color='green',label='prediction',linewidth=5)
plt.legend()
plt.show()
```
