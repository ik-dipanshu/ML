```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv(f"sample_data//california_housing_test.csv")
```

```python
from sklearn.preprocessing import MinMaxScaler
min_max_scaler = MinMaxScaler()
min_max_scaler.fit(df[['households']])
df['households']=min_max_scaler.transform(df[['households']])
df.head()
```

```python
sns.boxplot(x=df['households'])
plt.show()
```

```python
df=pd.DataFrame({'hours_of_study':[10,2,6,15,60],'marks':[80,20,40,87,95]})
df
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(df[['hours_of_study']],df['marks'])
model.predict([[50]])
```

```python
model.coef_
```

```python
model.intercept_
```

```python
plt.scatter(df['hours_of_study'],df['marks'])
plt.plot(df['hours_of_study'],model.predict(df[['hours_of_study']]),color='yellow')
plt.xlabel=('hours_of_study')
plt.ylabel=('marks')
plt .title('hours_of_study vs marks')
plt.show()
```
