```python
import pandas as pd

data = {
    'emp_id': [1,2,3,4,5,6,7,8],
    'emp_name': ['Amit','Neha','Ravi','Priya','Karan','Anjali','Suresh','Pooja'],
    'department': ['HR','IT','Finance','IT','HR','Finance','IT','HR'],
    'gender': ['Male','Female','Male','Female','Male','Female','Male','Female'],
    'education': ['Graduate','Postgraduate','Graduate','PhD','Postgraduate','Graduate','Graduate','PhD'],
    'city': ['Delhi','Mumbai','Pune','Delhi','Mumbai','Pune','Delhi','Mumbai'],
    'salary': [40000,60000,50000,80000,45000,52000,70000,48000],
    'experience': [2,5,3,6,1,4,8,9]
}

df = pd.DataFrame(data)
print(df)
```

```python
one_hotencode=pd.get_dummies(df['gender'],dtype=int,drop_first=True)
one_hotencode
```

```python
#using sklearn
from sklearn.preprocessing import OneHotEncoder
he=OneHotEncoder()
new_encoding=he.fit_transform(df[['gender']])
data=new_encoding.toarray()
df['encode_gender']=data[:,0]
```

```python
#label encoding
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['department']=le.fit_transform(df['department'])
```

```python
df
```

```python
#ordinal encoding
from sklearn.preprocessing import OrdinalEncoder
ord_enc=OrdinalEncoder()
df['education']=ord_enc.fit_transform(df[['education']])
df
```

```python
#linear regression
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
x=df['experience']
y=df['salary']
x=x.values.reshape(-1,1)
lr.fit(x,y)
```

```python
lr.predict([[5]])
```

```python
#how to use multilinear regression
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
x=df[['experience','education']]
y=df['salary']
lr.fit(x,y)
```

```python
#how to handle misssion records
import pandas as pd

data = {
    'emp_id': [1,2,3,4,5,6,7,8],
    'emp_name': ['Amit','Neha','Ravi','Priya','Karan','Anjali','Suresh','Pooja'],
    'department': ['HR',np.nan,'Finance','IT','HR','Finance','IT','HR'],
    'gender': ['Male','Female','Male','Female','Male','Female','Male','Female'],
    'education': ['Graduate','Postgraduate','Graduate','PhD','Postgraduate','Graduate','Graduate','PhD'],
    'city': ['Delhi','Mumbai','Pune','Delhi','Mumbai','Pune','Delhi','Mumbai'],
    'salary': [40000,np.nan,50000,80000,45000,52000,70000,48000],
    'experience': [2,5,3,6,1,4,8,9]
}

df = pd.DataFrame(data)
print(df)
```

```python
n=df['department'].mode()[0]
for i in range(len(df['department'])):
    x=df['department'].isnull()
    if x[i]==True:
        df['department'][i]=n

# df['department'].isnull()

```
