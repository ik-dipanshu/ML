```python
import matplotlib.pyplot as plt
```

```python
from matplotlib.lines import lineStyles
name=['ansh','pranjal','raghav','kuchbhi']
maths_marks=[90,50,99,56]
chem_marks=[60,50,39,58]
phy_marks=[10,80,93,86]

plt.plot(name,maths_marks,color='red',linestyle='-.',marker='s')
plt.plot(name,chem_marks,color='blue')
plt.plot(name,phy_marks,color='green')
plt.xlabel('name')
plt.ylabel('marks')
plt.title('marks of students')
plt.legend(['maths','chemistry','physics'])
plt.show()
```

Create a employeename and its salary graph keep labels and title

```python
plt.bar(['a','b','c','d','f'],[1,2,3,4,5],color='yellow',width=0.4,align='center',bottom=1000,edgecolor='blue',linewidth=4,alpha=0.4)
# alpha is use to change the transparency
plt.xlabel('name')
plt.ylabel('salary')
plt.title('title')
plt.show()
```

```python
plt.bar(['a','b','c','d','f'],[3,2,3,4,5],hatch='\\',label='kuchi bhi',)
plt.bar(['a','b','c','d','f'],[1,6,2,7,2],hatch='x',label='marketing',alpha=0.4)
#hatch use to insert design inside the bars ex- //, | ,\\ ,-,+,x,o
# alpha is use to change the transparency
plt.legend()
plt.xlabel('name')
plt.ylabel('salary')
plt.title('title')
plt.show()
```

```python
# error bar
plt.bar([1,2,3,4,5],['a','b','c','d','f'],xerr=[1,5,1,3,4],width=0.3,ecolor='red')
```

```python
plt.bar(['a','b','c','d','f'],[1,2,3,4,5],color='green',width=0.4,tick_label=['first','second','third','fourth','fifth'])
# alpha is use to change the transparency
#rotate tick to 90 degree
plt.xticks(rotation=45)
plt.yticks(rotation=270)
plt.xlabel('name')
plt.ylabel('salary')
plt.title('title')
plt.show()
```

```python
#pie chart
a=['A','B','C','D']
s=[15,30,45,10]
plt.pie(s,labels=a,colors=['red','yellow','green','blue'],autopct='%1.2f%%',explode=(0,0,0,0.3),shadow=True,startangle=180)
#startangle use to rotate your pie chart
plt.show()
```
